```python
## Global Variable -> variable that are created outside of a function
## Global variables can be used by everyone, both inside of functions and outside.

x = 'good boy'

def myfunc():
    print('priyanshu is ' + x)
    
myfunc()
```

```python
## If you create a variable with the same name inside a function, this variable will be local, and can only
# be used inside the function. The global variable with the same name will remain as it was, global and with
# the original value.

x = 'good boy'  # Global variables
def myfunc():
    x = 'very good boy'  # local variables
    print('priyanshu is ' + x)
    
myfunc()

print('priyasnshu is ' + x)
```

```python
## The global keyword
# Normally, when you create a variable inside a function, that variable is local, and can only be used inside that function.
# To create a global variable inside a function, you can use the global keyword.

def myfunc():
    global x
    x = ' best language'
    
myfunc()
print('python is ' + x)
```

```python
## To change the value of a global variable inside a function, refer to the variable by using the global keyword

x = 'fantastic'
def myfunc():
    global x
    x = ' best language'
    
myfunc()
print('python is ' + x)
```
