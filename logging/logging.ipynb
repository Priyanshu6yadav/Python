### Python Logging 
Python logging is used to record program events (info, warnings, errors, debug messages) instead of using print, helping with debugging, monitoring, and production tracking.

```python
# Python logging is used to track, debug, and monitor program execution using structured log messages.
import logging

# Configure the basic logging settings
logging.basicConfig(level=logging.DEBUG)

# Log message with different severity levels
logging.debug("This is a debug message")
logging.info("This is an info message")
logging.warning("This is a warning message")
logging.critical("This is a critical message")
```

#### Log Level
##### Log levels define message severity: DEBUG < INFO < WARNING < ERROR < CRITICAL.
	1	DEBUG – detailed information for debugging
	2.	INFO – general program flow messages
	3.	WARNING – something unexpected but program still runs
	4.	ERROR – program failed to perform a function
	5.	CRITICAL – severe error, program may stop


```python
# Configuring logging 
logging.basicConfig(
    level=logging.DEBUG,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt='%Y-%m-%d %H:%M:%S',
    force='True' ## for showing time and date and year
)
logging.debug("This is a debug message")
logging.info("This is an info message")
logging.warning("This is a warning message")
logging.critical("This is a critical message")
```

```python
# Configuring logging 
logging.basicConfig(
    filename='app.log',# filename='app.log' --> writes all log messages to a file named app.log instead of the console
    filemode='w', # filemode='w'--> opens the log file in write mode (overwrites the file each run)
    level=logging.DEBUG,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt='%Y-%m-%d %H:%M:%S',
    force='True' ## for showing time and date and year
)
logging.debug("This is a debug message")
logging.info("This is an info message")
logging.warning("This is a warning message")
logging.critical("This is a critical message")
```
