#### You can use an if inside another if ‚Äî this is called nested conditional.

```python
age = int(input("Enter your age: "))
if age >= 18:
    citizen = input("Are you a citizen if india? (yes/no): ")
    if citizen.lower()== "yes":
        print("You are eligible to vote")
    else:
        print("You are not eligible to vote.")
else:
    print("You are underage")
```

```python
## Student Result Evaluator
marks = int(input("Enter your marks: "))
if marks >=90:
    grade = "A+"
    remark = "Excellent"
elif marks >=75:
    grade = "A"
    remark = "Very Good"
elif marks >=60:
    grade = "B"
    remark = "Good"
elif marks >=40:
    grade = "C"
    remark = "Pass"
else:
    grade = "F"
    remark = "Fail"
print(f"Grade: {grade} | Remark: {remark}")
```

```python
# Q Write a program to check if a year is leap year or not.
# Conditions It is divisible by 400, It is divisible by 4 but not by 100.
year = int(input("Enter year"))
if (year % 400 ==0) or (year %4 ==0 and year %100 !=0):
    print(f"This is leap year {year}")
else:
    print("Not a leap year")
```

```python
## Write a program that takes 3 numbers and prints whether they form a right-angled triangle (using Pythagoras theorem).
A = int(input("Enter perpendicular value"))
B = int(input("Enter base value"))
C = int(input("Enter hypotenuse value"))
if C**2 == A**2 + B**2:
    print("It is a right angle triangle.")
elif C < A or C < B:
    print("Hypotenuse must be the largest side.")
else:
    print("It is not a right-angle triangle.")
```
