```python
# Default Parameters
def power(num, exp =2):
    return num ** exp
print(power(4))
print(power(4,3))
```

```python
## Multiple Return Values
def math_operations(a, b):
    add = a+b
    sub = a-b
    return add, sub
x, y= math_operations(10,4)
print("Sum:", x)
print("Difference: ", y)
```
